```python
import pandas as pd
from IPython import display # to refresh the screen
import time
import glob # to search each file to directory
import os # to get the current working directory

# initiate file name to use
file = ''

### Search excel in current directory
path = os.getcwd()

### Relative path
# for relative path please disable absolute path and enable relative path
# file = "./apple_2.xlsx"

### Absolute path
# for absolute path please disable relative path and enable absolute path
# file = "C:\Users\Hank\VS_code\"

# find alls xlsx files in the folder
excel_files = glob.glob(os.path.join(path, "*.xlsx"))
temp_filenames = []

for filename in excel_files:
    temp_filenames.append(filename.split('\\')[-1])

if len(temp_filenames) > 0:
    print("Which excel file that you want to use?\n")
    for num in range(0,len(temp_filenames)):
        print(f"{num + 1}.",temp_filenames[num])
    print("\n")
    try:
        file_choose = int(input("> "))
        if file_choose > 0 and file_choose <= len(temp_filenames):
            filename = temp_filenames[file_choose - 1]
            file = f"./{filename}"
            print(file)
    except:
        print("Wrong input, goodbye!")

else: 
    print("No excel file found!")



# initiate index variables
total_current_assets_index = 0
total_current_liabilities_index = 0
cash_index = 0
marketable_securities_index = 0
account_receivable_index = 0
accumulated_depreciation_index = 0
gross_ppe_index = 0
total_asset_index = 0
total_net_ppe_index = 0
total_inventories_index = 0
account_payable_index = 0
total_liabilities_index = 0
total_equity_index = 0
common_shares_outstanding_index = 0
total_revenue_index = 0
cogs_index = 0
ebit_index = 0
interest_expense_index = 0
net_income_index = 0
gross_proift_index = 0
operation_income_index = 0
rnd_expense_index = 0
cash_from_operating_activities_index = 0
max_year_limit = 0
min_year_limit = 0

# assign sheet names to variables
bs_data = pd.read_excel(io = file,
                     sheet_name= "Balance Sheet",
                     header=4
                     )
is_data = pd.read_excel(io = file,
                     sheet_name= "Income Statement",
                     header=4
                     )
cfs_data = pd.read_excel(io = file,
                     sheet_name= "Cash Flow",
                     header=4
                     )

# check index for each variables 
def GetAccountIndex():
    global total_current_assets_index
    global total_current_liabilities_index
    global cash_index
    global marketable_securities_index
    global account_receivable_index
    global accumulated_depreciation_index
    global gross_ppe_index
    global total_asset_index
    global total_net_ppe_index
    global total_inventories_index
    global account_payable_index
    global total_liabilities_index
    global total_equity_index
    global common_shares_outstanding_index
    global total_revenue_index
    global cogs_index
    global ebit_index
    global interest_expense_index
    global net_income_index
    global gross_proift_index
    global operation_income_index
    global rnd_expense_index
    global cash_from_operating_activities_index
    global min_year_limit
    global max_year_limit
    
    # use max and min year to limit the upper and lower limit of
    # user input in year
    bs_max_year = 0
    bs_min_year = 0
    is_max_year = 0
    is_min_year  =0
    cfs_max_year = 0
    cfs_min_year = 0

    bs_data_title = bs_data.iloc[:,0].astype(str)
    is_data_title = is_data.iloc[:,0].astype(str)
    cfs_data_title = cfs_data.iloc[:,0].astype(str)

    bs_max_year = int(bs_data.columns[1].split("-")[1])
    bs_min_year = int(bs_data.columns[-1].split("-")[1])
    is_max_year = int(is_data.columns[1].split("-")[1])
    is_min_year = int(is_data.columns[-1].split("-")[1])
    cfs_max_year = int(cfs_data.columns[1].split("-")[1])
    cfs_min_year = int(bs_data.columns[-1].split("-")[1])

    print("-" * 50)

    if bs_max_year == is_max_year and is_max_year == cfs_max_year:
        print(f"Most recent year : {bs_max_year}")
        max_year_limit = bs_max_year
    else:
        temp_year = [bs_max_year, is_max_year, cfs_max_year]
        print("*** Warning***\nYear in sheets are not consistance!\n"
              f"Most recent balance sheet year = {bs_max_year}\n"
              f"Most recent income statement year = {is_max_year}\n"
              f"Most recent cash flow statement year = {cfs_max_year}\n")
        max_year_limit = min(temp_year)
        
    if bs_min_year == is_min_year and is_min_year == cfs_min_year:
        print(f"Earliest year: {bs_min_year}")
        min_year_limit = bs_min_year
    else:
        temp_year = [bs_min_year, is_min_year, cfs_min_year]
        print("*** Warning***\nYear in sheets are not consistance!\n"
              f"Earliest recent balance sheet year = {bs_min_year}\n"
              f"Earliest recent income statement year = {is_min_year}\n"
              f"Earliest recent cash flow statement year = {cfs_min_year}\n")    
        min_year_limit = max(temp_year)

    print("-" * 50)

    # use for loop to iterate through each row and find the matching string
    # of chart account in balance sheet
    for row in range(0, len(bs_data_title)):
        if 'total current assets' in bs_data_title[row].lower():
            total_current_assets_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'total current liabilities' in bs_data_title[row].lower():
            total_current_liabilities_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'cash and cash equivalents' == bs_data_title[row].lstrip().lower():
            cash_index = row
            print(f'{bs_data_title[row].lstrip()}: Passed!')
        elif 'total short term investments' in bs_data_title[row].lower():
            marketable_securities_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'total trade and other receivables, current' in bs_data_title[row].lower():
            account_receivable_index = row
            print(f"{bs_data_title[row].lstrip()} (Account Receivable): Passed!")
        elif 'accumulated depreciation' == bs_data_title[row].lstrip().lower():
            accumulated_depreciation_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'total gross property, plant and equipment' in bs_data_title[row].lower():
            gross_ppe_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'total assets' == bs_data_title[row].lstrip().lower():
            total_asset_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'total net property, plant and equipment' in bs_data_title[row].lower():
            total_net_ppe_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!")
        elif 'total inventories' == bs_data_title[row].lstrip().lower():
            total_inventories_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!") 
        elif 'total payables and accrued expenses, current' in bs_data_title[row].lstrip().lower():
            account_payable_index = row
            print(f"{bs_data_title[row].lstrip()} (Account Payable): Passed!") 
        elif 'total liabilities' == bs_data_title[row].lstrip().lower():
            total_liabilities_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!") 
        elif 'total equity' == bs_data_title[row].lstrip().lower():
            total_equity_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!") 
        elif 'common shares outstanding' in bs_data_title[row].lstrip().lower():
            common_shares_outstanding_index = row
            print(f"{bs_data_title[row].lstrip()}: Passed!") 
        else:
            pass

    # use for loop to iterate through each row and find the matching string
    # of chart account in income statement
    for row in range(0, len(is_data_title)):
        if 'total revenue' == is_data_title[row].lstrip().lower():
            total_revenue_index = row
            print(f"{is_data_title[row].lstrip()}: Passed!")
        elif 'total cost of revenue' == is_data_title[row].lstrip().lower():
            cogs_index = row
            print(f"{is_data_title[row].lstrip()} (COGS): Passed!")
        elif 'pretax income' == is_data_title[row].lstrip().lower():
            ebit_index = row
            print(f"{is_data_title[row].lstrip()} (EBIT): Passed!")
        elif 'interest expense net of capitalized interest' in is_data_title[row].lstrip().lower():
            interest_expense_index = row
            print(f"{is_data_title[row].lstrip()} (Interest Expense): Passed!")
        elif 'net income from continuing operations' in is_data_title[row].lstrip().lower():
            net_income_index = row
            print(f"{is_data_title[row].lstrip()} (Net Income): Passed!")
        elif 'total gross profit' == is_data_title[row].lstrip().lower():
            gross_proift_index = row
            print(f"{is_data_title[row].lstrip()}: Passed!")
        elif 'total operating profit/(loss)' in is_data_title[row].lstrip().lower():
            operation_income_index = row
            print(f"{is_data_title[row].lstrip()} (Operating Income): Passed!")
        elif 'research and development expenses' in is_data_title[row].lstrip().lower():
            rnd_expense_index = row
            print(f"{is_data_title[row].lstrip()} (R&D Expense): Passed!")
        else:
            pass

    # use for loop to iterate through each row and find the matching string
    # of chart account in cash flow statement
    for row in range(0,len(cfs_data_title)):
        if 'total cash flow from operating activities, indirect' in cfs_data_title[row].lstrip().lower():
            cash_from_operating_activities_index = row
            print(f"{cfs_data_title[row].lstrip()}: Passed!")
        else:
            pass

# Current Ratio function iterate through the each column and match the user input year
def CurrentRatio(year):
    for col in bs_data:
        if year in col:
            current_asset = bs_data.loc[total_current_assets_index,col]
            current_liabilities = bs_data.loc[total_current_liabilities_index, col]
            current_ratio = round(current_asset/current_liabilities,2)
            print(f"Current Assets: {current_asset:,}\n"
                   f"Current Liabilities: {current_liabilities:,}")
            print('-' * 50)
            print(f"Current Ration for year {year} is ",current_ratio)
            break

# Acid Test function iterate through the each column and match the user input year
def AcidTest(year):
    for col in bs_data:
        if year in col:
            cash = bs_data.loc[cash_index,col]
            account_receivable = bs_data.loc[account_receivable_index,col]
            current_liabilities = bs_data.loc[total_current_liabilities_index, col]
            marketable_secutities = bs_data.loc[marketable_securities_index, col]
            acid_test_ratio = round((cash + account_receivable + marketable_secutities)/current_liabilities,2)
            print(f"Cash: {cash:,}\n"
                  f"Account Receivable: {account_receivable:,}\n"
                  f"Marketable Securities: {marketable_secutities:,}\n"
                  f"Current Liabilities: {current_liabilities:,}")
            print('-' * 50)
            print(f"Acid Test Ration for year {year} is ",acid_test_ratio)
            break

# Asset turn over ratio iterate through the each column and match the user input year
# total asset is average so it take average of the prior year and the current year 
# fort the calculation
def AssetTurnover(year):
    sale = 0
    y1_total_assets = 0
    y2_total_assets =0
    avg_total_assets = 0
    asset_turnover = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_total_assets = bs_data.loc[total_asset_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_total_assets = bs_data.loc[total_asset_index,col]
                break
        for col in is_data:
            if year in col:
                sale = is_data.loc[total_revenue_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_total_assets = round((y1_total_assets + y2_total_assets) / 2,2)
    asset_turnover = round(sale/avg_total_assets,2)

    print(f"Revenue: {sale:,}")
    print(f"Average Total Assets: {avg_total_assets:,}")
    print('-' * 50)
    print(f"Asset Turnover Ratio for {year} is ", asset_turnover)

    # for index test
    # print("2022: ", y1_total_assets)
    # print("2021: ", y2_total_assets)
    # print("avg: ",avg_total_assets)
    # print("sale:", sale)

# PPE turn over ratio iterate through the each column and match the user input year
# total net ppe is average so it take average of the prior year and the current year 
# fort the calculation
def PPETurnover(year):
    sale = 0
    y1_net_ppe = 0
    y2_net_ppe =0
    avg_net_ppe = 0
    ppe_turnover = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_net_ppe = bs_data.loc[total_net_ppe_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_net_ppe = bs_data.loc[total_net_ppe_index,col]
                break
        for col in is_data:
            if year in col:
                sale = is_data.loc[total_revenue_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_net_ppe = round((y1_net_ppe + y2_net_ppe) / 2,2)
    ppe_turnover = round(sale/avg_net_ppe,2)

    print(f"Revenue: {sale:,}")
    print(f"Average Net PPE: {avg_net_ppe:,}")
    print('-' * 50)
    print(f"PPE Turnover Ratio for {year} is ", ppe_turnover)

# Account Receivable turn over ratio iterate through the each column and match the user input year
# Account Receivable is average so it take average of the prior year and the current year 
# fort the calculation
def ARTurnover(year):
    sale = 0
    y1_acc_rec = 0
    y2_acc_rec =0
    avg_acc_rec = 0
    acc_rec_turnover = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_acc_rec = bs_data.loc[total_net_ppe_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_acc_rec = bs_data.loc[total_net_ppe_index,col]
                break
        for col in is_data:
            if year in col:
                sale = is_data.loc[total_revenue_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_acc_rec = round((y1_acc_rec + y2_acc_rec) / 2,2)
    acc_rec_turnover = round(sale/avg_acc_rec,2)

    print(f"Revenue: {sale:,}")
    print(f"Average Account Receivable: {avg_acc_rec:,}")
    print('-' * 50)
    print(f"Account Receivable Turnover Ratio for {year} is ", acc_rec_turnover)

# Inventory turn over ratio iterate through the each column and match the user input year
# Inventory is average so it take average of the prior year and the current year 
# fort the calculation
def InventoryTurnover(year):
    cogs = 0
    y1_inventory = 0
    y2_inventory =0
    avg_inventory = 0
    inventory_turnover = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_inventory = bs_data.loc[total_inventories_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_inventory = bs_data.loc[total_inventories_index,col]
                break
        for col in is_data:
            if year in col:
                cogs = is_data.loc[cogs_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_inventory = round((y1_inventory + y2_inventory) / 2,2)
    inventory_turnover = round(cogs/avg_inventory,2)

    print(f"Revenue: {cogs:,}")
    print(f"Average Inventory: {avg_inventory:,}")
    print('-' * 50)
    print(f"Inventory Turnover Ratio for {year} is ", inventory_turnover)

# Account Payable turn over ratio iterate through the each column and match the user input year
# Account Payable is average so it take average of the prior year and the current year 
# fort the calculation
def APTurnover(year):
    cogs = 0
    y1_acc_payable = 0
    y2_acc_payable =0
    avg_acc_payable = 0
    acc_payable_turnover = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_acc_payable = bs_data.loc[account_payable_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_acc_payable = bs_data.loc[account_payable_index,col]
                break
        for col in is_data:
            if year in col:
                cogs = is_data.loc[cogs_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_acc_payable = round((y1_acc_payable + y2_acc_payable) / 2,2)
    acc_payable_turnover = round(cogs/avg_acc_payable,2)

    print(f"Revenue: {cogs:,}")
    print(f"Average Account Payable: {avg_acc_payable:,}")
    print('-' * 50)
    print(f"Account Payable Turnover Ratio for {year} is ", acc_payable_turnover)

# Percent PPE Used Up ratio iterate through the each column and match the user input year
def PercentPPEUsedUp(year):
    accumulate_depreciation = 0
    gross_ppe = 0
    percent_ppe_used_up = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                accumulate_depreciation = bs_data.loc[accumulated_depreciation_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                gross_ppe = bs_data.loc[gross_ppe_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    percent_ppe_used_up = round(accumulate_depreciation/gross_ppe,2)

    print(f"Accumulated Depreciation: {accumulate_depreciation:,}")
    print(f"Gross PPE: {gross_ppe:,}")
    print('-' * 50)
    print(f"Percent PPE Used Up Ratio for {year} is ", percent_ppe_used_up)

# Days Sales Outstanding ratio iterate through the each column and match the user input year
# Account Receivable is average so it take average of the prior year and the current year 
# fort the calculation
def DSO(year):
    sale = 0
    y1_acc_rec = 0
    y2_acc_rec =0
    avg_acc_rec = 0
    dso = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_acc_rec = bs_data.loc[total_net_ppe_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_acc_rec = bs_data.loc[total_net_ppe_index,col]
                break
        for col in is_data:
            if year in col:
                sale = is_data.loc[total_revenue_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_acc_rec = round((y1_acc_rec + y2_acc_rec) / 2,2)
    dso = round(avg_acc_rec/(sale/365),2)
    
    print(f"Average Account Receivable: {avg_acc_rec:,}")
    print(f"Revenue: {sale:,}")
    print('-' * 50)
    print(f"Days Sales Outstanding (DSO) for {year} is ", dso)

# Days In Inventory ratio iterate through the each column and match the user input year
# Inventory is average so it take average of the prior year and the current year 
# fort the calculation
def DIO(year):
    cogs = 0
    y1_inventory = 0
    y2_inventory =0
    avg_inventory = 0
    dio = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_inventory = bs_data.loc[total_inventories_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_inventory = bs_data.loc[total_inventories_index,col]
                break
        for col in is_data:
            if year in col:
                cogs = is_data.loc[cogs_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_inventory = round((y1_inventory + y2_inventory) / 2,2)
    dio = round(avg_inventory/(cogs/365),2)

    print(f"Average Inventory: {avg_inventory:,}")
    print(f"Revenue: {cogs:,}")
    print('-' * 50)
    print(f"Days in Inventory (DIO) for {year} is ", dio)

# Days Payable Outstanding ratio iterate through the each column and match the user input year
# Account Payable is average so it take average of the prior year and the current year 
# fort the calculation
def DPO(year):
    cogs = 0
    y1_acc_payable = 0
    y2_acc_payable =0
    avg_acc_payable = 0
    dpo = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_acc_payable = bs_data.loc[account_payable_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_acc_payable = bs_data.loc[account_payable_index,col]
                break
        for col in is_data:
            if year in col:
                cogs = is_data.loc[cogs_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_acc_payable = round((y1_acc_payable + y2_acc_payable) / 2,2)
    dpo = round(avg_acc_payable/(cogs/365),2)

    print(f"Average Account Payable: {avg_acc_payable:,}")
    print(f"Revenue: {cogs:,}")
    print('-' * 50)
    print(f"Days Payable Outstanding (DPO) for {year} is ", dpo)

# Cash Conversion Cycle ratio iterate through the each column and match the user input year
# Account Receivable is average so it take average of the prior year and the current year 
# fort the calculation
# This is the calculation of DSO, DIO, and DPO
def CCC(year):
    cogs = 0
    y1_acc_payable = 0
    y2_acc_payable =0
    avg_acc_payable = 0
    sale = 0
    y1_acc_rec = 0
    y2_acc_rec =0
    avg_acc_rec = 0
    y1_inventory = 0
    y2_inventory =0
    avg_inventory = 0
    y1_acc_payable = 0
    y2_acc_payable =0
    avg_acc_payable = 0
    dpo = 0
    dio = 0
    dso = 0
    ccc = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_acc_payable = bs_data.loc[account_payable_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_acc_payable = bs_data.loc[account_payable_index,col]
                break
        for col in is_data:
            if year in col:
                cogs = is_data.loc[cogs_index,col]
                break
        for col in bs_data:
            if str(year) in col:
                y1_acc_rec = bs_data.loc[total_net_ppe_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_acc_rec = bs_data.loc[total_net_ppe_index,col]
                break
        for col in is_data:
            if year in col:
                sale = is_data.loc[total_revenue_index,col]
                break
        for col in bs_data:
            if str(year) in col:
                y1_inventory = bs_data.loc[total_inventories_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_inventory = bs_data.loc[total_inventories_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_acc_payable = round((y1_acc_payable + y2_acc_payable) / 2,2)
    dpo = round(avg_acc_payable/(cogs/365),2)
    avg_acc_rec = round((y1_acc_rec + y2_acc_rec) / 2,2)
    dso = round(avg_acc_rec/(sale/365),2)
    avg_inventory = round((y1_inventory + y2_inventory) / 2,2)
    dio = round(avg_inventory/(cogs/365),2)
    ccc = dso + dio - dpo

    print(f"Average Account Receivable: {avg_acc_rec:,}")
    print(f"Revenue: {sale:,}")
    print(f"Average Inventory: {avg_inventory:,}")
    print(f"Revenue: {cogs:,}")
    print(f"Average Account Payable: {avg_acc_payable:,}")
    print('-' * 50)
    print(f"Days Sales Outstanding (DSO) for {year} is ", dso)
    print(f"Days in Inventory (DIO) for {year} is ", dio)
    print(f"Days Payable Outstanding (DPO) for {year} is ", dpo)
    print('-' * 50)
    print(f"Cash Conversion Cycle (CCC) for {year} is ", ccc)

# Liabilities to Equity ratio iterate through the each column and match the user input year
def LiabilitiesToEquity(year):
    for col in bs_data:
        if year in col:
            total_equity = bs_data.loc[total_equity_index,col]
            total_liabilities = bs_data.loc[total_liabilities_index, col]
            liabilities_to_equity_ratio = round(total_liabilities/total_equity,2)
            print(f"Total Liabilities: {total_liabilities:,}")
            print(f"Total Assets: {total_equity:,}")
            print('-' * 50)
            print(f"Liabilities to Equity Ration for year {year} is ",liabilities_to_equity_ratio)
            break

# Financial Levarage as standalone ratio iterate through the each column and match the user input year
def FinancialLeverageStandalone(year):
    for col in bs_data:
        if year in col:
            total_equity = bs_data.loc[total_equity_index,col]
            total_assets = bs_data.loc[total_asset_index, col]
            financial_leverage_standalone = round(total_assets/total_equity,2)
            print(f"Total Liabilities: {total_assets:,}")
            print(f"Total Assets: {total_equity:,}")
            print('-' * 50)
            print(f"Financial Leverage as a standalone ration for year {year} is ",financial_leverage_standalone)
            break    

# Financial Leverage (for decompisition) ratio iterate through the each column and match the user input year
# Total assets and total equity are average so it take average of the prior year and the current year 
# for the calculation
def FinancialLeverageDecomposition(year):
    y1_total_assets = 0
    y2_total_assets = 0
    avg_total_assets = 0
    y1_total_equity = 0
    y2_total_equity = 0
    avg_total_equity = 0
    financial_leverage_decomposition = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_total_assets = bs_data.loc[total_asset_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_total_assets = bs_data.loc[total_asset_index,col]
                break
        for col in bs_data:
            if str(year) in col:
                y1_total_equity = bs_data.loc[total_equity_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_total_equity = bs_data.loc[total_equity_index,col]
                break

    else:
        print("Sorry wrong input, please try again!")

    avg_total_assets = round((y1_total_assets + y2_total_assets) / 2,2)
    avg_total_equity = round((y1_total_equity + y2_total_equity) / 2,2)
    financial_leverage_decomposition = round(avg_total_assets/avg_total_equity,2)

    print(f"Average Total Liabilities: {avg_total_assets:,}")
    print(f"Average Total Assets: {avg_total_equity:,}")
    print('-' * 50)
    print(f"Financial Leverage (for decomposition) ration for year {year} is ",financial_leverage_decomposition)

# Time Interest Earned ratio iterate through the each column and match the user input year
def TimesInterestEarned(year):
    for col in is_data:
        if year in col:
            ebit = is_data.loc[ebit_index,col]
            interest_exp = is_data.loc[interest_expense_index, col]
            times_interest_earned = round(interest_exp/ebit,2)
            print(f"Earning Before Interest and Taxes: {ebit:,}")
            print(f"Interest Expense: {interest_exp:,}")
            print('-' * 50)
            print(f"Times Interest Earned Ratio for year {year} is ",times_interest_earned)
            break   

# Cash from operation by Debt ratio iterate through the each column and match the user input year
# Total Debt is average so it take average of the prior year and the current year 
# for the calculation
def CashFromOperationByDebt(year):
    cash_from_opt = 0
    y1_total_debt = 0
    y2_total_debt = 0
    avg_total_debt = 0
    cash_from_opt_by_debt = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in cfs_data:
            if str(year) in col:
                cash_from_opt = cfs_data.loc[cash_from_operating_activities_index,col]
                break 
        for col in bs_data:
            if str(year) in col:
                y1_total_debt = bs_data.loc[total_liabilities_index,col]
                break 
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_total_debt = bs_data.loc[total_liabilities_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")

    avg_total_debt = round((y1_total_debt + y2_total_debt) / 2,2)
    cash_from_opt_by_debt = round(cash_from_opt/avg_total_debt,2)

    print(f"Cash from Operating Activities: {cash_from_opt:,}")
    print(f"Average Total Debt: {avg_total_debt:,}")
    print('-' * 50)
    print(f"Cash from Operating Activities/ Average Total Debt Ratio for year {year} is ",cash_from_opt_by_debt)
      
# Return on Asset ratio iterate through the each column and match the user input year
# Total Asset is average so it take average of the prior year and the current year 
# for the calculation
def ROA(year):
    net_income = 0
    y1_total_assets = 0
    y2_total_assets =0
    avg_total_assets = 0
    roa = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_total_assets = bs_data.loc[total_asset_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_total_assets = bs_data.loc[total_asset_index,col]
                break
        for col in is_data:
            if year in col:
                net_income = is_data.loc[net_income_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_total_assets = round((y1_total_assets + y2_total_assets) / 2,2)
    roa = round(net_income/avg_total_assets,2)

    print(f"Net Income: {net_income:,}")
    print(f"Average Total Assets: {avg_total_assets:,}")
    print('-' * 50)
    print(f"Return on Assets (ROA) Ratio for {year} is ", roa)

# Return on Equity ratio iterate through the each column and match the user input year
# Total equity is average so it take average of the prior year and the current year 
# for the calculation
def ROE(year):
    net_income = 0
    y1_total_equity = 0
    y2_total_equity =0
    avg_total_equity = 0
    roe = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_total_equity = bs_data.loc[total_equity_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_total_equity = bs_data.loc[total_equity_index,col]
                break
        for col in is_data:
            if year in col:
                net_income = is_data.loc[net_income_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_total_equity = round((y1_total_equity + y2_total_equity) / 2,2)
    roe = round(net_income/avg_total_equity,2)

    print(f"Net Income: {net_income:,}")
    print(f"Average Total Equity: {avg_total_equity:,}")
    print('-' * 50)
    print(f"Return on Equity (ROE) Ratio for {year} is ", roe)

# Gross Profit Margin ratio iterate through the each column and match the user input year
def GrossProfitMargin(year):
    for col in is_data:
        if year in col:
            gross_profit = is_data.loc[gross_proift_index,col]
            sale = is_data.loc[total_revenue_index, col]
            gross_profit_margin = round(gross_profit/sale,2)
            print(f"Gross Profit: {gross_profit:,}")
            print(f"Revenue (Sale): {sale:,}")
            print('-' * 50)
            print(f"Gross Profit Margin Ratio for year {year} is ",gross_profit_margin)
            break   

# Operating Profit Margin ratio iterate through the each column and match the user input year
def OptProfitMargin(year):
    for col in is_data:
        if year in col:
            opt_income = is_data.loc[operation_income_index,col]
            sale = is_data.loc[total_revenue_index, col]
            opt_profit_margin = round(opt_income/sale,2)
            print(f"Operating Income: {opt_income:,}")
            print(f"Revenue (Sale): {sale:,}")
            print('-' * 50)
            print(f"Operating Profit Margin Ratio for year {year} is ",opt_profit_margin)
            break  
    
# Net Profiting Margin ratio iterate through the each column and match the user input year
def NetProfitMargin(year):
    for col in is_data:
        if year in col:
            net_income = is_data.loc[net_income_index,col]
            sale = is_data.loc[total_revenue_index, col]
            net_profit_margin = round(net_income/sale,2)
            print(f"Net Income: {net_income:,}")
            print(f"Revenue (Sale): {sale:,}")
            print('-' * 50)
            print(f"Net Profit Margin Ratio for year {year} is ",net_profit_margin)
            break  

# Earnings per share ratio iterate through the each column and match the user input year
# Common Share Outstanding is average so it take average of the prior year and the current year 
# for the calculation
def EPS(year):
    net_income = 0
    y1_common_share_oustnd = 0
    y2_common_share_oustnd =0
    avg_common_share_oustnd = 0
    eps = 0
    if int(year) <= max_year_limit and int(year) >= min_year_limit:
        for col in bs_data:
            if str(year) in col:
                y1_common_share_oustnd = bs_data.loc[common_shares_outstanding_index,col]
                break
        for col in bs_data:
            if str(int(year)-1) in col:
                y2_common_share_oustnd = bs_data.loc[common_shares_outstanding_index,col]
                break
        for col in is_data:
            if year in col:
                net_income = is_data.loc[net_income_index,col]
                break
    else:
        print("Sorry wrong input, please try again!")
    
    avg_common_share_oustnd = round((y1_common_share_oustnd + y2_common_share_oustnd) / 2,2)
    eps = round(net_income/avg_common_share_oustnd,2)

    print(f"Net Income: {net_income:,}")
    print(f"Average Common Share Outstanding: {avg_common_share_oustnd:,}")
    print('-' * 50)
    print(f"Earnings Per Share (EPS) Ratio for {year} is ", eps)

# Research and Dev. to Percent of Sale ratio iterate through the each column and match the user input year
def RNDAsPercentOfSale(year):
    if rnd_expense_index == 0:
        print("Sorry no R&D expense.")
    else:
        for col in is_data:
            if year in col:
                rnd = is_data.loc[rnd_expense_index,col]
                sale = is_data.loc[total_revenue_index, col]
                rnd_as_percent_of_sales = round(rnd/sale,2)
                print(f"Research and Development (R&D) Expense: {rnd:,}")
                print(f"Revenue (Sale): {sale:,}")
                print('-' * 50)
                print(f"R&D as a percent of sales ratio for year {year} is ",rnd_as_percent_of_sales)
                break  

# Main function consolidate call of all the functions in the program.
# show a list of calculation, upon user input with year return the calcualtion accordingly.
def main():
    yes_list = ["yes", "yeah", "y", "please", "ok", "sure", "go", "do"]
    no_list = ["no", "nah", "nope", "don't", "not", "cancel", "reject", "exit"]
    print("\n\n Do you want to continue?")
    user_input = input("> ")
    time.sleep(1)
    display.clear_output()
    if user_input.lower() in yes_list:
        while True:
            print("Please choose the ratios:\n\n"
                "--------------- Liquidity Ratios ---------------\n"
                "1. Current Ratio\n"
                "2. Quick or Acid Test Ratios")
            print('-' * 14, "Productivity Ratio", '-' * 14)
            print("3. Asset Turnover Ratio\n"
                "4. PPE Turnover\n"\
                "5. Account Receivable Turnover\n"
                "6. Inventory Turnover\n"
                "7. Account Payable Turnover\n"
                "8. Percent PPE Used Up")
            print('-' * 12, "Cash Conversion Cycle", '-' * 13)
            print("9. Days Sales Outstanding (DSO)\n"
                "10. Days in Inventory (DIO)\n"\
                "11. Days Payable Outstanding (DPO)\n"
                "12. Cash Conversion Cycle (CCC)")
            print('-' * 11, "Financial Leverage Ratio", '-' * 11)
            print("13. Liabilities to Equity\n"
                "14. Financial Levarage as a standalone ratio\n"
                "15. Financial Levarage (for decomposition)\n"
                "16. Times Interest Earned\n"
                "17. Cash from Operations/Total Debt")
            print('-' * 13, "Profitability Ratios", '-' * 13)
            print("18. Return on Assets (ROA)\n"
                "19. Return on Equity (ROE)\n"
                "20. Gross Profit Margin\n"
                "21. Operating Profit Margin\n"
                "22. Net Profit Margin")
            print('-' * 10, "Other Commonly Used Ratios", '-' * 10)
            print("23. Earning Per Share (EPS)\n"
                "24. Research and Development as a Percent of Sales")
            user_input = input("> \n")
            print("\n")
            
            if user_input == '1':
                print("This is the Current Ratio calculation. Please choose a year: ")
                year = input("> \n")
                CurrentRatio(year)
                print("\n")

            elif user_input == '2':
                print("This is the Quick or Acid Test Ratio calculation. Please choose a year: ")
                year = input("> \n")
                AcidTest(year)
                print("\n")

            elif user_input =='3':
                print("This is the Assets Turnover Ratio calculation. Please choose a year: ")
                year = input("> \n")
                AssetTurnover(year)
                print("\n")

            elif user_input =='4':
                print("This is the PPE Turnover Ratio calculation. Please choose a year: ")
                year = input("> \n")
                PPETurnover(year)
                print("\n")

            elif user_input =='5':
                print("This is the Account Receivable Turnover Ratio calculation. Please choose a year: ")
                year = input("> \n")
                ARTurnover(year)
                print("\n")

            elif user_input =='6':
                print("This is the Inventory Turnover Ratio calculation. Please choose a year: ")
                year = input("> \n")
                InventoryTurnover(year)
                print("\n")

            elif user_input =='7':
                print("This is the Account Payable Turnover Ratio calculation. Please choose a year: ")
                year = input("> \n")
                APTurnover(year)
                print("\n")

            elif user_input =='8':
                print("This is the Percent PPE Used Up Ratio calculation. Please choose a year: ")
                year = input("> \n")
                PPETurnover(year)
                print("\n")

            elif user_input =='9':
                print("This is the Days Sales Outstanding (DSO) calculation. Please choose a year: ")
                year = input("> \n")
                DSO(year)
                print("\n")

            elif user_input =='10':
                print("This is the Days in Inventory (DIO) calculation. Please choose a year: ")
                year = input("> \n")
                DIO(year)
                print("\n")

            elif user_input =='11':
                print("This is the Days Payable Outstanding (DPO) calculation. Please choose a year: ")
                year = input("> \n")
                DPO(year)
                print("\n")

            elif user_input =='12':
                print("This is the Cash Conversion Cycle (CCC) calculation. Please choose a year: ")
                year = input("> \n")
                CCC(year)
                print("\n")

            elif user_input =='13':
                print("This is the Liabilities to Equity Ratio calculation. Please choose a year: ")
                year = input("> \n")
                LiabilitiesToEquity(year)
                print("\n")

            elif user_input =='14':
                print("This is the Financial Leverage as a Standalone Ratio calculation. Please choose a year: ")
                year = input("> \n")
                FinancialLeverageStandalone(year)
                print("\n")

            elif user_input =='15':
                print("This is the Financial Leverage (for decomposition) Ratio calculation. Please choose a year: ")
                year = input("> \n")
                FinancialLeverageDecomposition(year)
                print("\n")

            elif user_input =='16':
                print("This is the Times Interest Earned Ratio calculation. Please choose a year: ")
                year = input("> \n")
                TimesInterestEarned(year)
                print("\n")
            
            elif user_input =='17':
                print("This is the Cash from Operating Activities / Average Total Debt Ratio calculation. Please choose a year: ")
                year = input("> \n")
                CashFromOperationByDebt(year)
                print("\n")

            elif user_input =='18':
                print("This is the Return on Assets (ROA) Ratio calculation. Please choose a year: ")
                year = input("> \n")
                ROA(year)
                print("\n")

            elif user_input =='19':
                print("This is the Return on Equity (ROE) Ratio calculation. Please choose a year: ")
                year = input("> \n")
                ROE(year)
                print("\n")

            elif user_input =='20':
                print("This is the Gross Profit Margin Ratio calculation. Please choose a year: ")
                year = input("> \n")
                GrossProfitMargin(year)
                print("\n")

            elif user_input =='21':
                print("This is the Operating Profit Margin Ratio calculation. Please choose a year: ")
                year = input("> \n")
                OptProfitMargin(year)
                print("\n")

            elif user_input =='22':
                print("This is the Net Profit Margin Ratio calculation. Please choose a year: ")
                year = input("> \n")
                NetProfitMargin(year)
                print("\n")

            elif user_input =='23':
                print("This is the Earnings Per Share (EPS) Ratio calculation. Please choose a year: ")
                year = input("> \n")
                EPS(year)
                print("\n")

            elif user_input =='24':
                print("This is the R&D as a percent of sales ratio calculation. Please choose a year: ")
                year = input("> \n")
                RNDAsPercentOfSale(year)
                print("\n")


            elif user_input.lower() == 'exit' or user_input.lower() == 'quit':
                break

            else:
                print("I don't understand. Please try again!")
                print("\n")

    # bs_data.loc[total_asset_index,"Dec-2020"]

    elif user_input.lower() in no_list:
        print("Ok good bye!")

    else:
        print("Sorry, I don't understand you, goodbye")



# Check the variable names and get indexes
GetAccountIndex()

# Run the calculation program
main()
```
